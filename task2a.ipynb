```python
new_df = data['country'].value_counts().rename_axis('countries').reset_index(name='counts')
new_df.head(3)
```

```python
your_labels = new_df.countries
your_values = new_df.counts
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
ax.pie(your_values, labels = your_labels, autopct='%1.2f%%')
plt.title("countries distribution")
plt.show()
```

for the above piechart i have used the autopct which is used to calculate the percentage.France has the more percentage than the other 2 counties.

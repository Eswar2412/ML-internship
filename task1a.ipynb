```python
# Histogram for numerical variables
data['credit_score'].hist(bins=20)
plt.title('Histogram of Numerical Column')
plt.xlabel('Numerical Column')
plt.ylabel('Frequency')
plt.show()
```

The resulting plot is a histogram that visualizes the distribution of 'credit_score' values in the dataset. Each bar in the histogram represents the frequency of values falling within a specific range  of credit scores

```python
# Countplot for categorical variables
sns.countplot(x='country', data=data)
plt.title('Countplot of Categorical Column')
plt.xlabel('country')
plt.ylabel('Count')
plt.show()
```

displays the count of observations

```python
# Scatter plot for numerical variables
sns.scatterplot(x='age', y='credit_score', data=data)
plt.title('Scatter Plot of Numerical Columns')
plt.xlabel('age')
plt.ylabel('credit_score')
plt.show()
```

It is a type of data visualization that displays individual data points on a two-dimensional graph.

```python
# Box plot for numerical and categorical variables
sns.boxplot(x='country', y='age', data=data)
plt.title('Box Plot of Numerical Column by Categorical Column')
plt.xlabel('country')
plt.ylabel('age')
plt.show()
```

```python
# Correlation matrix
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pairplot for selected variables
sns.pairplot(data[['age', 'credit_score', 'estimated_salary']])
#plt.title('Pairplot of Selected Numerical Columns')
plt.show()
```

```python
#Box plot for outlier detection
sns.boxplot(x=data['credit_score'])
plt.title('Box Plot for Outlier Detection')
plt.xlabel('Numerical Column')
plt.show()
```
